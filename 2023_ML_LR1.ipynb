```python
def totuple(a):
    try:
        return tuple(totuple(i) for i in a)
    except TypeError:
        return a
```

```python
def grid(n):
    x = np.linspace(0, 1, n)
    y = np.linspace(0, 1, n)
    xx, yy = np.meshgrid(x, y)
    grid = np.dstack((xx, yy))
    return grid.reshape(n**2,2)
```

```python
print(grid(3))
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_grid(n):
    gr = grid(n).reshape(n,n,2)
    x_coords = gr[:, :, 0].flatten()
    y_coords = gr[:, :, 1].flatten()

    plt.scatter(x_coords, y_coords)
    plt.show()

plot_grid(3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML


M = 101
random_matrix = np.random.rand(M, M)
X = (random_matrix < 0.1).astype(int)
fig, ax = plt.subplots(figsize=(6, 6))

def update(frame):
    global X
    N = (
        np.roll(X, 1, 1) + np.roll(X, -1, 1) +
        np.roll(X, 1, 0) + np.roll(X, -1, 0) +
        np.roll(np.roll(X, 1, 0), 1, 1) + np.roll(np.roll(X, -1, 0), -1, 1) +
        np.roll(np.roll(X, -1, 1), 1, 0)  +np.roll(np.roll(X, 1, 1), -1, 0) 
    )    
    X = ((N == 3) | (X & (N == 2))).astype(int)
    ax.clear()
    ax.imshow(X, cmap='binary')
    ax.set_title(f'Шаг {frame}')

ani = FuncAnimation(fig, update, frames=range(1, M), interval=200, repeat=False)

#plt.show()
HTML(ani.to_jshtml())
```

```python
def ExtendedGCD(a, b):
    M = np.eye(2)
    while b:
        q = a // b
        a, b = b, a % b
        M = np.dot(M, np.array([[0,1],[1, -q]]))
    return M[0,0].astype(int), M[1,0].astype(int)
```

Модель линейной регрессии в самом простом виде предполагает следующее.

Пусть заданы два набора числовых данных $X=\{x_1,x_2,...x_k\}$ и $Y=\{y_1,y_2,...y_k\}$. Необходимо по известным данным отыскать такую функцию вида $f(x) = a*x +b $ максимально описывающую зависимость представителей набора $Y$ от представителей набора $X$.

Задача обучения в данном случае состоит в нахождении коэффициентов $a$ и $b$ таких, чтобы минимизировать значение квадратичной ошибки $SSE(a,b) = \displaystyle\sum_{i=1}^{k}(y_i - f(x_i))^2$ на исходных данных. 

Разрешив полученную задачу оптимизации можно получить формулы для коэффициентов $a$ и $b$.

$$a = \frac{\displaystyle\sum_{i=1}^{k}(x_k - \overline{x})(y_k - \overline{y})}{\displaystyle\sum_{i=1}^{k}(x_k - \overline{x})^2}$$

$$b = \overline{y}- a \overline{x} $$

Полученные формулы можно представить в **матричном виде**, если записать исходные данные в вектор столбцы или вектор строки.

$$a = \frac{(X - \overline{x})(Y - \overline{y})}{(X - \overline{x})^2},$$ где $X=\begin {pmatrix} x_1 \\ \vdots \\ x_k \end {pmatrix}$ и $Y=\begin{pmatrix}y_1, y_2, \dots y_k\end {pmatrix}$.

$$b = \overline{y}- a \overline{x} $$

```python
np.shape(y)
```

```python
def linear_regression(X, Y):
    mean_x = np.mean(X)
    mean_y = np.mean(Y)
    a = np.dot((X - mean_x), (Y - mean_y)) / np.dot(X - mean_x, X - mean_x)
    b = mean_y - a * mean_x 
    return a, b
```

```python
a, b = linear_regression(x, y)
```
